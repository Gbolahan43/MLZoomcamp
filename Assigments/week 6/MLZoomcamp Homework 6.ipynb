```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
len(df_train), len(df_test), len(df_val)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeRegressor, export_text
from sklearn.metrics import roc_auc_score
```

```python
dv = DictVectorizer(sparse=True)
```

```python
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_train.head()
```

```python
y_train = df_train['jamb_score'].values
del df_train['jamb_score']
y_test = df_test['jamb_score'].values
del df_test['jamb_score']
y_val = df_val['jamb_score'].values
del df_val['jamb_score']
```

```python
train_dict = df_train.to_dict(orient= 'records')

X_train = dv.fit_transform(train_dict)

val_dict = df_val.to_dict(orient= 'records')

X_val = dv.transform(val_dict)

test_dict = df_test.to_dict(orient= 'records')

test_val = dv.transform(val_dict)
```

```python
dt = DecisionTreeRegressor(max_depth=1)
```

```python
dt.fit(X_train, y_train)
```

```python
dv.get_feature_names_out()
```

```python
print(export_text(dt, feature_names=dv.get_feature_names_out()))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=10,
                           random_state=1,
                           n_jobs=-1)
```

```python
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_val)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(y_val, rf_pred)
```

```python
scores = []
for n in range(10,201,10):
    rf = RandomForestRegressor(n_estimators=n,
                           random_state=1,
                           n_jobs=-1)
    rf.fit(X_train, y_train)
    rf_pred = rf.predict(X_val)
    rmse = round(root_mean_squared_error(y_val, rf_pred), 3)
    scores.append((n, rmse))
```

```python
scores
```

```python
scores = []
max_depths= [10, 15, 20, 25]
for m in max_depths:
    for n in range(10,201,10):
        rf = RandomForestRegressor(n_estimators=n,
                                   max_depth=m,
                                   random_state=1,
                                n_jobs=-1)
        rf.fit(X_train, y_train)
        rf_pred = rf.predict(X_val)
        rmse = round(root_mean_squared_error(y_val, rf_pred), 3)
        scores.append((n, m, rmse))
```

```python
scores_df = pd.DataFrame(scores, columns=['n_est', 'max_depth', 'rmse'])
```

```python
scores_df.sort_values('rmse', ascending=True)
```
